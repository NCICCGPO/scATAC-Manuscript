```r
library(ggseqlogo)
library(ggplot2)
library(rtracklayer)
library(BSgenome.Hsapiens.UCSC.hg38)
library(patchwork) 
library(reticulate) 
```

```r
cs1 = make_col_scheme(chars=c('A', 'C', 'G', 'T'), 
                      cols=c('#109648', '#255C99', '#F7B32B', '#D62839'))
```

```r
plot_seq <- function(m, ymin=0, ymax=2, title=NULL, method='custom') {
    rownames(m) = c("A","C","G","T")
    
    p <- ggseqlogo(m, method=method, ncol=1,col_scheme=cs1) + 
                                    ylim(c(ymin,ymax)) 
    
    if (!is.null(title)) {
        p = p + ggtitle(title)
        p = p + theme(plot.title = element_text(hjust = 0.5, size=18))
    }
    
    p
}
```

```r
plot_npy_shaps <- function(shap_path, seq_width, shift, title, fig_save_path, plot_negatives=TRUE) {

    np<-import("numpy") #datareading 
    mat<-np$load(shap_path) 
    dim(mat)

    len <- 1364

    ind_start <- as.integer((len - seq_width)/2) + shift
    ind_end <- ind_start + seq_width - 1
    ind_start
    ind_end

    mat1 <- t(drop(mat))[,ind_start:ind_end]
    dimnames(mat1) <- list(c('A', 'C', 'G', 'T'))
    mat1
    
    ymax <- max(mat1)
    ymin <- 0
    if (plot_negatives) {
        ymin <- -ymax
    }

    p = plot_seq(mat1, ymin=ymin, ymax=ymax, title=title)
    ggsave(p, file=fig_save_path, height=2, width=20, useDingbats=F)
    p
}
```

```r
shifts <- c(71, 44, 51, 78, 2, 57, 0, 14, 81, 61, 33, 83, 10, 50, 61, 78, 67)
shifts
```

```r
shift_idx = 1
for (x in append(c(10:25),"")) {
    shift <- shifts[shift_idx]
    for (plot_negatives in c(TRUE,FALSE)) {
        sample_name <- sprintf("BRCA%s",x)
        print(sample_name)
        print(plot_negatives)

        shap_path <- sprintf("/illumina/scratch/deep_learning/akumar22/TCGA/vignette/brca_vignette_figure/%s_scores.npy", sample_name)
        shap_path

        seq_width <- 200
        shift

        title <- sample_name

        fig_save_path <- sprintf("/illumina/scratch/deep_learning/akumar22/TCGA/vignette/brca_R_plots/%s_%s_%s_R.pdf", sample_name, seq_width, plot_negatives)
        fig_save_path


        plot_npy_shaps(shap_path, seq_width, shift, title, fig_save_path, plot_negatives)
        
        }
    shift_idx <- shift_idx + 1
}
```
